```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import matplotlib as mpl

import seaborn as sns
sns.set()

%matplotlib inline
```

```python

mpl.rcParams['axes.titlesize'] = 16
mpl.rcParams['axes.labelsize'] = 16
mpl.rcParams['xtick.labelsize'] = 13
mpl.rcParams['ytick.labelsize'] = 13
```

```python
russia_df = pd.read_csv('tweets.csv', index_col = 'user_key')
```

```python
russia_df.sort_values('retweet_count', ascending=False).head(10)
```

```python
russia_df['created_datetime'] = pd.to_datetime(russia_df['created_str'])
russia_df['created_datetime'] = russia_df['created_datetime'] - pd.Timedelta(hours=5)
```

```python
del russia_df['created_str']
```

```python
russia_df.set_index('created_datetime', inplace=True)
```

```python
russia_df.sample(5)
```

```python
russia_df['text'].sample(5)
```

```python
plt.figure(figsize=(15,5))
hourly_tweets = russia_df.resample('M')

counts_per_hour = hourly_tweets['posted'].count()


#plt.plot(counts_per_hour.index.month, counts_per_hour)
busiest_window_start = '2016-6-1'
busiest_window_end = '2017-7-1'
counts_per_hour.plot()

plt.axvspan(busiest_window_start, busiest_window_end, facecolor='red', alpha=0.5)
plt.xlabel('Tweet Month')
plt.ylabel('Number of Tweets')
plt.title('Monthly Russian Troll Tweets')

plt.figure(figsize=(15,5))


busiest_window = russia_df.loc[busiest_window_start:busiest_window_end]

daily_tweets = busiest_window.resample('D')

counts_per_day = daily_tweets['posted'].count()
total_RTs_per_hour = daily_tweets['retweet_count'].sum()
hourly_window_start = '2016-10-6'
hourly_window_end = '2016-10-7'

#counts_per_day.plot()
total_RTs_per_hour.plot()
plt.axvspan(hourly_window_start, hourly_window_end, facecolor='red', alpha=0.5)
plt.xlabel('Tweet Date')
plt.ylabel('Number of Tweets')
plt.title('Daily Russian Troll Tweets (June 2016 - July 2017)');

plt.figure(figsize=(15,5))

busy_month_sample = russia_df.loc[hourly_window_start:hourly_window_end]
daily_tweets = busy_month_sample.resample('H')

counts_per_day = daily_tweets['posted'].count()

#plt.plot(counts_per_hour.index.month, counts_per_hour)
counts_per_day.plot()
plt.xlabel('Tweet Hour')
plt.ylabel('Number of Tweets')
plt.title('Hourly Russian Troll Tweets (Oct 6 2017 - Oct 7 2017)');
```
